# 4. If Statements (Conditionals)

Introduce the concept of decision-making using if, else, and elif.

```python
# Basic if-else example
number = int(input('Enter a number: >>> '))

if number > 0:
  print("You are positive!")
  if number % 2 == 0:
    print("You are even!")
  else:
      print("You are odd!")
else:
  print("You are negative!")
```
